```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()

G.add_edge('a', 'b', weight=0.6)
G.add_edge('a', 'c', weight=0.2)
G.add_edge('c', 'd', weight=0.1)
G.add_edge('c', 'e', weight=0.7)
G.add_edge('c', 'f', weight=0.9)
G.add_edge('a', 'd', weight=0.3)

elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.5]
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] <= 0.5]

pos = nx.spring_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# edges
nx.draw_networkx_edges(G, pos, edgelist=elarge,
                       width=6)
nx.draw_networkx_edges(G, pos, edgelist=esmall,
                       width=6, alpha=0.7, edge_color='b', style='dashed')

# labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

plt.axis('off')
plt.show()
```

```python
G = nx.Graph()
G.add_node(0, label=r'$R_{x}$', pos=(1, 1) )
G.add_node(4, label=r'$R_{y}$', pos=(1, 2)  )
G.add_node(1, label=r'$R_{y}$', pos=(2, 2)  )
G.add_node(2, label=r'$R_{z}$', pos=(3, 2)  )
G.add_node(3, label=r'$R_{x,z}$', pos=(2, 3)  )
G.add_node(5, label=r'$R_{x,z}$', pos=(1.5, 2)  )

G.add_edge(0,1,weight=0.00003)
G.add_edge(1,2, weight=45)
G.add_edge(1,3, weight=20)
G.add_edge(0,3, weight=0.020)
G.add_edge(2,3, weight=200)
G.add_edge(0,4, weight=200)


pos=nx.get_node_attributes(G,'pos')
labels = nx.get_node_attributes(G, 'label')
edges,weights = zip(*nx.get_edge_attributes(G,'weight').items())
```
