# MODEL LEARNING TEST: Offline Data

### Preliminaries

```python
spinlst = [evo.sigmax(), evo.sigmay(), evo.sigmaz()]
```

```python
eoplist = list(map(lambda j: np.kron(spinlst[j],np.eye(2)) , range(len(spinlst))))

HFoplist = list(map(lambda j: np.kron(spinlst[j],spinlst[j]), range(len(spinlst))))
bathoplist = list(map(lambda j: np.kron(np.eye(2),spinlst[j]) , range(len(spinlst))))
```

```python
TwoQlist = []
TwoQlist.extend(HFoplist)
TwoQlist.extend(bathoplist)
```

```python
TwoQsample = []
TwoQsample.append(HFoplist[0])
TwoQsample.append(HFoplist[2])
TwoQsample.append(bathoplist[1])
```

```python
HahnTestingList = [np.array([eoplist[2]]), np.array([HFoplist[1]]), 
                   np.array(eoplist), np.array(HFoplist), np.array(TwoQsample), 
                   np.array(TwoQlist)]
```

### Launch the learning

```python
allnameslist = ['1q_1p', '1q_full', '2qHF_1p', '2qHF_full', '2q_advanced', '2q_full']
```

```python
modeltest = qmd.ModelsDevelopmentClass(HahnTestingList, allnameslist, checkloss=True, gaussian = False, trotter=False, IQLE=False, datasource='offline')

print("The expected true model is: " + str(modeltest.TrueNames))
```

```python
directory = os.path.join("..","..","..","Hamiltonian Learning With Seb","Decoherence_learning","NV05_HahnEcho02")
moredata =  mIO.ImportAllFolder_Hahnpeak(directory, clean_duplicates = True)  

myrange = range(0, min(moredata.shape[0],500)) 

# the 1000 factor converts ns to us
moredata = [ [moredata[i,0]/1000 , moredata[i,1]] for i in myrange]
moredata = np.array(moredata)

offset = 0.18
moredata[:,0] = moredata[:,0]-offset

expdata = mIO.rescaledatatofullrange(moredata[:,1])
```

```python
myHfA = FinEstimParams['2q_full'][range(3),0]
myBathfreq = FinEstimParams['2q_full'][range(3),1]
```

```python
datarange = len(SETTINGS.mydata) #select the data range for which the run has been performed
datamean = np.mean(expdata[0:datarange])
datavar = np.sum( (expdata[0:datarange] -datamean)**2  )
print(datavar)

datarange = 109
simdata = HTM.Q2_HF_Precession(moredata[0:datarange,0], HfA = myHfA, Bathfreq = myBathfreq)
simRsq = np.around(( 1 -  np.sum( (simdata - expdata[0:datarange])**2  )/datavar   ), 2)

priorHfA = np.array([-2.7, -2.7, -2.14])
priorBathfreq = np.array([3., 1., 1.])
priordata = HTM.Q2_HF_Precession(moredata[0:datarange,0], HfA = priorHfA, Bathfreq = priorBathfreq)
priorRsq = np.around(( 1 -  np.sum( (priordata - expdata[0:datarange])**2  )/datavar   ), 2)

print("Prior model R^2 = " + str(priorRsq))
print("Learnt model R^2 = " + str(simRsq))
```

```python
tvec = np.linspace(0.,1.6,200)

simdata = HTM.Q2_HF_Precession(tvec, HfA = myHfA, Bathfreq = myBathfreq)
priordata = HTM.Q2_HF_Precession(tvec, HfA = priorHfA, Bathfreq = priorBathfreq)
```

```python
plt.plot(tvec, simdata, "b-", label='Learnt Model')
plt.plot(tvec, priordata, "g-", label = "Initial Model")

plt.plot(moredata[:,0], expdata, "r.", label='Exp. data')

plt.xlim([0,1.6])
plt.legend()

plt.text(0.82, 0.15, r'R$^2$ (learnt) = ' + str(simRsq) )
plt.text(0.82, 0.05, r'R$^2$ (initial) = ' + str(priorRsq) )

plt.title(r'Hahn Echo peaks - no Deco')
plt.xlabel(r'time $\tau_1 (\mu s)$')
plt.ylabel(r'Pr0 = $\langle \Psi | 0 \rangle ^2 $')

filefigname = dire+"\CompareData_LearntModel_noDeco"+mytimestamp+".pdf"
filefigname = os.path.normpath(filefigname)
plt.savefig(filefigname)
print('Picture saved in\n'+filefigname)

plt.show()
```

# MULTIPLE testing running

```python
nruns = 10
pool_len = 20
max_expnum = 30

collectTrueOpList = []
collectTrueNames = []
collectTrueParams = []

collectEstimParams = []

collectKfactorsDict = []
collectLogL = []


for run in range(nruns):
    modeltest = qmd.ModelsDevelopmentClass(HahnTestingList, allnameslist, checkloss=True, gaussian = False, trotter=False, IQLE=False, experimental='offline')
    collectTrueNames.append(modeltest.TrueNames)
    
    modeltest.InitialiseAllActiveModels()
    
    collectTrueParams.append(modeltest.TrueParamsList)
    
    start=time.clock()
    modeltest.UpdateAllActiveModels(expnum=max_expnum)
    end=time.clock()
    print('Total elapsed time for the learning process: ' + str(end-start))
    
    collectEstimParams.append([modeltest.ModelsList[i].NewEval for i in range(len(modeltest.ModelsList)) ])
    
    collectKfactorsDict.append(modeltest.ComputeAllBayesFactors(fromLogL = True))
    
    collectLogL.append([modeltest.ModelsList[i].KLogTotLikelihood for i in range(len(modeltest.ModelsList)) ])
    
```

## Data export

```python
import pickle
PIK = "testQMD_QLE_epochs"+str(max_expnum)+"pool"+str(pool_len)+".dat"
```

```python
data = [collectTrueOpList, collectTrueNames, collectTrueParams, collectEstimParams, collectKfactorsDict, collectLogL]
with open(PIK, "wb") as f:
    pickle.dump(data, f)
```
